```python
for _, tutor in df_tutors.iterrows():
    tutor_id = tutor["tutor_id"]
    preferred_mode = tutor["preferred_mode"]
    max_hours = tutor["max_weekly_hours"]

    for day in weekdays:
        for block in time_blocks:

            start_hour = int(block.split(":")[0])

            # bazowa logika dostępności
            if tutor_id == 5:
                # studentka - mniej godzin, głównie popołudnia i wieczory
                available = 1 if day in ["Monday", "Wednesday", "Friday", "Saturday"] and start_hour >= 15 else 0

            elif tutor_id == 3:
                # online tutor - głównie popołudnia, bez sobotnich poranków
                available = 1 if start_hour >= 14 and day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] else 0

            elif tutor_id == 4:
                # in-person tutor - bardziej klasyczny grafik
                available = 1 if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"] and 10 <= start_hour <= 18 else 0

            else:
                # mixed tutors
                available = 1 if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"] and 12 <= start_hour <= 20 else 0

            tutor_availability.append({
                "availability_id": availability_id,
                "tutor_id": tutor_id,
                "weekday": day,
                "time_block": block,
                "available_flag": available
            })

            availability_id += 1

df_tutor_availability = pd.DataFrame(tutor_availability)
df_tutor_availability.head(20)
```

```python
def choose_tutor(student_row, tutors_df):
    age_group = student_row["age_group"]
    learning_goal = student_row["learning_goal"]
    assigned_mode = student_row["assigned_mode"]

    # kopiujemy tutorów do filtrowania
    possible_tutors = tutors_df.copy()

    # 1. tryb
    if assigned_mode == "online":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["online", "mixed"])]
    else:
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["in_person", "mixed"])]

    if len(filtered) > 0:
        possible_tutors = filtered

    # 2. exam prep częściej do bardziej doświadczonych
    if age_group == "exam_prep" or learning_goal in ["matura_basic", "matura_extended"]:
        filtered = possible_tutors[possible_tutors["experience_level"].isin(["mid", "senior"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 3. adults częściej do online / mixed
    if age_group == "adult":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["online", "mixed"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 4. dzieci częściej do mixed / in-person
    if age_group == "child":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["in_person", "mixed"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 5. tutor 5 ma mniej uczniów
    tutor_ids = possible_tutors["tutor_id"].tolist()

    weights_map = {
        1: 1.2,
        2: 1.0,
        3: 0.9,
        4: 1.1,
        5: 0.5
    }

    weights = [weights_map[tutor_id] for tutor_id in tutor_ids]

    chosen_tutor_id = random.choices(tutor_ids, weights=weights, k=1)[0]

    return chosen_tutor_id
```

```python
assigned_tutor_ids = []

for _, student in df_students.iterrows():
    tutor_id = choose_tutor(student, df_tutors)
    assigned_tutor_ids.append(tutor_id)

df_students["assigned_tutor_id"] = assigned_tutor_ids
df_students.head()
```

```python
df_tutor_available["slot_key"] = df_tutor_available["weekday"] + "|" + df_tutor_available["time_block"]
df_student_available["slot_key"] = df_student_available["weekday"] + "|" + df_student_available["time_block"]
```

```python
def assign_weekly_slots(student_row, df_student_available, df_tutor_available):
    student_id = student_row["student_id"]
    tutor_id = student_row["assigned_tutor_id"]
    lessons_per_week = student_row["lessons_per_week"]

    student_slots = set(
        df_student_available[df_student_available["student_id"] == student_id]["slot_key"]
    )

    tutor_slots = set(
        df_tutor_available[df_tutor_available["tutor_id"] == tutor_id]["slot_key"]
    )

    common_slots = list(student_slots.intersection(tutor_slots))

    if len(common_slots) == 0:
        return []

    # sortowanie
    weekday_order = {
        "Monday": 1,
        "Tuesday": 2,
        "Wednesday": 3,
        "Thursday": 4,
        "Friday": 5,
        "Saturday": 6
    }

    common_slots_sorted = sorted(
        common_slots,
        key=lambda x: (
            weekday_order[x.split("|")[0]],
            x.split("|")[1]
        )
    )

    # los - tyle slotów, ile trzeba
    if len(common_slots_sorted) <= lessons_per_week:
        selected_slots = common_slots_sorted
    else:
        selected_slots = random.sample(common_slots_sorted, lessons_per_week)

    return selected_slots
```

```python
weekly_slots = []

for _, student in df_students.iterrows():
    selected_slots = assign_weekly_slots(student, df_student_available, df_tutor_available)

    for slot in selected_slots:
        weekday, time_block = slot.split("|")

        weekly_slots.append({
            "student_id": student["student_id"],
            "assigned_tutor_id": student["assigned_tutor_id"],
            "weekday": weekday,
            "time_block": time_block,
            "assigned_mode": student["assigned_mode"],
            "typical_duration_minutes": student["typical_duration_minutes"],
            "lessons_per_week": student["lessons_per_week"]
        })

df_weekly_slots = pd.DataFrame(weekly_slots)
df_weekly_slots.head(20)
```

```python
df_lessons = df_lessons.sort_values(["lesson_date", "time_block", "tutor_id", "student_id"]).reset_index(drop=True)
df_lessons["lesson_id"] = range(1, len(df_lessons) + 1)

df_lessons.head()
```

```python
student_columns = [
    "student_id",
    "student_name",
    "age_group",
    "cefr_level",
    "learning_goal",
    "assigned_tutor_id",
    "assigned_tutor_name",
    "prep_profile",
    "sibling_group_id"
]

df_lessons = df_lessons.merge(
    df_students[student_columns],
    on="student_id",
    how="left"
)

df_lessons.head()
```

```python
tutor_columns = [
    "tutor_id",
    "tutor_name",
    "experience_level",
    "hourly_rate",
    "preferred_mode",
    "home_area"
]

df_lessons = df_lessons.merge(
    df_tutors[tutor_columns],
    on="tutor_id",
    how="left"
)

df_lessons.head()
```

```python
statuses = []

for _, row in df_lessons.iterrows():
    month = row["month"]

    # wyższe ryzyko anulowania zimą
    if month in [12, 1, 2]:
        status = random.choices(
            ["completed", "cancelled_by_student", "cancelled_by_tutor", "no_show"],
            weights=[84, 10, 2, 4]
        )[0]
    else:
        status = random.choices(
            ["completed", "cancelled_by_student", "cancelled_by_tutor", "no_show"],
            weights=[88, 7, 2, 3]
        )[0]

    statuses.append(status)

df_lessons["status"] = statuses

df_lessons["status"].value_counts()
```

```python
prep_times = []

for _, row in df_lessons.iterrows():
    prep_profile = row["prep_profile"]

    if prep_profile == "low":
        prep_time = random.choice([5, 10])
    elif prep_profile == "medium":
        prep_time = random.choice([10, 15])
    else:
        prep_time = random.choice([15, 20])

    prep_times.append(prep_time)

df_lessons["prep_time_minutes"] = prep_times

df_lessons["prep_time_minutes"].value_counts().sort_index()
```

```python
travel_times = []

for _, row in df_lessons.iterrows():
    if row["mode"] == "in_person":
        travel_time = random.choice([10, 15, 20, 25, 30, 35, 40])
    else:
        travel_time = 0

    travel_times.append(travel_time)

df_lessons["travel_time_minutes"] = travel_times

df_lessons["travel_time_minutes"].value_counts().sort_index()
```

```python
platforms = []

for _, row in df_lessons.iterrows():
    if row["mode"] == "online":
        platform = random.choices(["Teams", "Skype", "Zoom"], weights=[50, 35, 15])[0]
    else:
        platform = "in_person"

    platforms.append(platform)

df_lessons["platform"] = platforms

df_lessons["platform"].value_counts()
```

```python
df_lessons["base_price"] = (df_lessons["hourly_rate"] * df_lessons["duration_minutes"] / 60).round(2)

df_lessons[["hourly_rate", "duration_minutes", "base_price"]].head()
```

```python
discount_types = []
final_prices = []

for _, row in df_lessons.iterrows():
    base_price = row["base_price"]
    sibling_group_id = row["sibling_group_id"]
    duration = row["duration_minutes"]

    discount_type = "none"
    discount = 0.0

    if pd.notna(sibling_group_id):
        discount_type = "sibling_discount"
        discount = 0.10

    if duration >= 90:
        if discount_type == "none":
            discount_type = "long_lesson_discount"
            discount = 0.08
        else:
            discount_type = "sibling_and_long_lesson"
            discount = 0.15

    final_price = round(base_price * (1 - discount), 2)

    discount_types.append(discount_type)
    final_prices.append(final_price)

df_lessons["discount_type"] = discount_types
df_lessons["final_price"] = final_prices

df_lessons[["base_price", "discount_type", "final_price"]].head(10)
```

```python
for _, row in df_lessons.iterrows():
    lesson_id = row["lesson_id"]
    lesson_date = row["lesson_date"]
    status = row["status"]
    amount_due = row["final_price"]

    if status == "completed":
        payment_status = random.choices(
            ["paid", "late", "missing"],
            weights=[88, 9, 3]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "late":
            amount_paid = amount_due
        else:
            amount_paid = 0

    elif status == "cancelled_by_student":
        payment_status = random.choices(
            ["paid", "partial", "missing"],
            weights=[20, 35, 45]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "partial":
            amount_paid = round(amount_due * 0.5, 2)
        else:
            amount_paid = 0

    elif status == "cancelled_by_tutor":
        payment_status = "waived"
        amount_paid = 0

    else:  # no_show
        payment_status = random.choices(
            ["paid", "partial", "missing"],
            weights=[45, 25, 30]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "partial":
            amount_paid = round(amount_due * 0.5, 2)
        else:
            amount_paid = 0

    payment_rows.append({
        "lesson_id": lesson_id,
        "lesson_date": lesson_date,
        "amount_due": amount_due,
        "amount_paid": amount_paid,
        "payment_status": payment_status
    })

df_payments = pd.DataFrame(payment_rows)
df_payments.head()
```

```python
df_payments["payment_id"] = range(1, len(df_payments) + 1)
df_payments.head()
```

```python
payment_dates = []

for _, row in df_payments.iterrows():
    lesson_date = pd.to_datetime(row["lesson_date"])
    payment_status = row["payment_status"]

    if payment_status == "paid":
        offset_days = random.choice([0, 1, 2, 3])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    elif payment_status == "late":
        offset_days = random.choice([5, 7, 10, 14])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    elif payment_status == "partial":
        offset_days = random.choice([0, 2, 5, 7])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    else:
        payment_date = pd.NaT

    payment_dates.append(payment_date)

df_payments["payment_date"] = payment_dates
df_payments.head()
```

```python
tutor_revenue = df_lessons.groupby(["tutor_id", "tutor_name"]).agg(
    expected_revenue=("final_price", "sum")
).reset_index()

actual_by_tutor = df_lessons.merge(
    df_payments[["lesson_id", "amount_paid"]],
    on="lesson_id",
    how="left"
).groupby(["tutor_id", "tutor_name"]).agg(
    actual_revenue=("amount_paid", "sum")
).reset_index()

tutor_revenue_summary = tutor_revenue.merge(actual_by_tutor, on=["tutor_id", "tutor_name"])
tutor_revenue_summary["revenue_gap"] = (
    tutor_revenue_summary["expected_revenue"] - tutor_revenue_summary["actual_revenue"]
).round(2)

tutor_revenue_summary.sort_values("expected_revenue", ascending=False)
```

```python
age_group_summary = df_lessons.groupby("age_group").agg(
    total_lessons=("lesson_id", "count"),
    avg_price=("final_price", "mean"),
    total_revenue=("final_price", "sum")
).reset_index()

age_group_summary.sort_values("total_revenue", ascending=False)
```

```python
mode_summary = df_lessons.groupby("mode").agg(
    total_lessons=("lesson_id", "count"),
    avg_price=("final_price", "mean"),
    total_revenue=("final_price", "sum"),
    avg_travel_time=("travel_time_minutes", "mean")
).reset_index()

mode_summary
```

```python
payment_summary = df_payments.groupby("payment_status").agg(
    count=("payment_id", "count"),
    total_due=("amount_due", "sum"),
    total_paid=("amount_paid", "sum")
).reset_index()

payment_summary
```

```python
monthly_summary = df_lessons.groupby(["year", "month", "month_name"]).agg(
    total_lessons=("lesson_id", "count"),
    expected_revenue=("final_price", "sum")
).reset_index()

monthly_actual = df_lessons.merge(
    df_payments[["lesson_id", "amount_paid"]],
    on="lesson_id",
    how="left"
).groupby(["year", "month", "month_name"]).agg(
    actual_revenue=("amount_paid", "sum")
).reset_index()

monthly_summary = monthly_summary.merge(
    monthly_actual,
    on=["year", "month", "month_name"],
    how="left"
)

monthly_summary["revenue_gap"] = (
    monthly_summary["expected_revenue"] - monthly_summary["actual_revenue"]
).round(2)

monthly_summary.sort_values(["year", "month"])
```
