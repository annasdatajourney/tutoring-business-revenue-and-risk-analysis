# Key Business Insights

## Revenue
- Expected revenue: ~707k
- Actual revenue: ~631k
- Revenue leakage: ~75k (~11%)

## Main Issue
- 11% revenue loss driven by:
  - student cancellations
  - no-shows

## Root Cause
- Problem is not tutors
- Problem is student behaviour

## High-Risk Groups
- Teens → highest revenue loss
- Pre-teens → second highest
- Adults → lowest risk

## Seasonality
- Highest losses:
  - January
  - December
  - February

## Business Recommendations
- Introduce prepayment
- Add cancellation policy
- Use reminders (SMS/email)
- Implement subscription model
