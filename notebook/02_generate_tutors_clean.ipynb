## 3. Sibling Groups

This section assigns some students to sibling groups.

The purpose is to simulate realistic family-based enrolment patterns, where more than one student may come from the same household.

This matters because sibling groups can influence:
- lesson scheduling
- retention patterns
- household-level pricing decisions

```python
df_students = df_students.copy()

df_students["sibling_group_id"] = pd.NA

num_students = len(df_students)
target_share = 0.20  # around 20% of students have siblings
target_count = int(num_students * target_share)

available_ids = df_students["student_id"].tolist()
random.shuffle(available_ids)

group_id = 1
assigned_count = 0
idx = 0

while assigned_count < target_count and idx < len(available_ids) - 1:
  # most groups have 2 students, occasionally 3
    group_size = random.choices([2, 3], weights=[85, 15])[0]

    group_members = available_ids[idx:idx + group_size]

    if len(group_members) < 2:
        break

    df_students.loc[df_students["student_id"].isin(group_members), "sibling_group_id"] = group_id

    assigned_count += len(group_members)
    idx += group_size
    group_id += 1

df_students.head(10)
```

## 4. Tutor Availability

This section generates tutor availability across weekdays and time blocks.

Availability is modelled to reflect realistic working patterns, taking into account:
- preferred teaching days
- time-of-day preferences
- maximum weekly hours

This creates realistic capacity constraints for later scheduling analysis.

```python
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
time_blocks = [
    "08:00-09:00",
    "09:00-10:00",
    "10:00-11:00",
    "11:00-12:00",
    "12:00-13:00",
    "13:00-14:00",
    "14:00-15:00",
    "15:00-16:00",
    "16:00-17:00",
    "17:00-18:00",
    "18:00-19:00",
    "19:00-20:00",
    "20:00-21:00"
]

tutor_availability = []
availability_id = 1
```

```python
for _, tutor in df_tutors.iterrows():
    tutor_id = tutor["tutor_id"]
    preferred_mode = tutor["preferred_mode"]
    max_hours = tutor["max_weekly_hours"]

    for day in weekdays:
        for block in time_blocks:

            start_hour = int(block.split(":")[0])

            # base availability logic
            if tutor_id == 5:
                # student tutor - fewer hours, mainly afternoons and evenings
                available = 1 if day in ["Monday", "Wednesday", "Friday", "Saturday"] and start_hour >= 15 else 0

            elif tutor_id == 3:
                # online tutor - mainly afternoons, no Saturday mornings
                available = 1 if start_hour >= 14 and day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] else 0

            elif tutor_id == 4:
                # in-person tutor - more traditional schedule
                available = 1 if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"] and 10 <= start_hour <= 18 else 0

            else:
                # mixed-mode tutors
                available = 1 if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"] and 12 <= start_hour <= 20 else 0

            tutor_availability.append({
                "availability_id": availability_id,
                "tutor_id": tutor_id,
                "weekday": day,
                "time_block": block,
                "available_flag": available
            })

            availability_id += 1

df_tutor_availability = pd.DataFrame(tutor_availability)
df_tutor_availability.head(20)
```

## Student Availability

This section generates student availability based on age group and typical daily routines.

Availability patterns reflect realistic behaviour observed in a tutoring business:
- younger students are mostly available after school
- older students and exam preparation groups prefer later hours
- adults tend to prefer evening slots

This helps simulate real scheduling constraints when matching students with tutors.

```python
student_availability = []
availability_id = 1
```

```python
for _, student in df_students.iterrows():
    student_id = student["student_id"]
    age_group = student["age_group"]

    for day in weekdays:
        for block in time_blocks:
            start_hour = int(block.split(":")[0])

            # default: not available
            available = 0

            if age_group == "child":
                # children: mostly after school, some Saturday availability
                if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] and 15 <= start_hour <= 18:
                    available = 1 if random.random() < 0.75 else 0
                elif day == "Saturday" and 9 <= start_hour <= 13:
                    available = 1 if random.random() < 0.50 else 0

            elif age_group == "pre_teen":
                # pre-teens: after school plus slightly later hours
                if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] and 15 <= start_hour <= 19:
                    available = 1 if random.random() < 0.70 else 0
                elif day == "Saturday" and 9 <= start_hour <= 14:
                    available = 1 if random.random() < 0.45 else 0

            elif age_group == "teen":
                # teens: afternoons and evenings
                if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] and 16 <= start_hour <= 20:
                    available = 1 if random.random() < 0.68 else 0
                elif day == "Saturday" and 10 <= start_hour <= 15:
                    available = 1 if random.random() < 0.35 else 0

            elif age_group == "exam_prep":
                # exam preparation students: later hours, often also Saturdays
                if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] and 17 <= start_hour <= 20:
                    available = 1 if random.random() < 0.72 else 0
                elif day == "Saturday" and 10 <= start_hour <= 16:
                    available = 1 if random.random() < 0.40 else 0

            elif age_group == "adult":
                # adults: mainly evenings, occasionally mornings or Saturdays
                if day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"] and 18 <= start_hour <= 20:
                    available = 1 if random.random() < 0.75 else 0
                elif day == "Saturday" and 9 <= start_hour <= 13:
                    available = 1 if random.random() < 0.30 else 0

            student_availability.append({
                "availability_id": availability_id,
                "student_id": student_id,
                "weekday": day,
                "time_block": block,
                "available_flag": available
            })

            availability_id += 1

df_student_availability = pd.DataFrame(student_availability)
df_student_availability.head(20)
```

## Tutor Assignment

This section assigns tutors to students based on selected business rules.

The matching logic takes into account:
- lesson mode
- student age group
- learning goal
- tutor experience
- tutor availability patterns

This helps simulate more realistic tutor-student matching within the business model.

```python
def choose_tutor(student_row, tutors_df):
    age_group = student_row["age_group"]
    learning_goal = student_row["learning_goal"]
    assigned_mode = student_row["assigned_mode"]

    possible_tutors = tutors_df.copy()

    # 1. lesson mode
    if assigned_mode == "online":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["online", "mixed"])]
    else:
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["in_person", "mixed"])]

    if len(filtered) > 0:
        possible_tutors = filtered

    # 2. exam preparation students are more often assigned to more experienced tutors
    if age_group == "exam_prep" or learning_goal in ["matura_basic", "matura_extended"]:
        filtered = possible_tutors[possible_tutors["experience_level"].isin(["mid", "senior"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 3. adults are more often assigned to online or mixed-mode tutors
    if age_group == "adult":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["online", "mixed"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 4. children are more often assigned to mixed-mode or in-person tutors
    if age_group == "child":
        filtered = possible_tutors[possible_tutors["preferred_mode"].isin(["in_person", "mixed"])]
        if len(filtered) > 0:
            possible_tutors = filtered

    # 5. tutor 5 is assigned fewer students
    tutor_ids = possible_tutors["tutor_id"].tolist()

    weights_map = {
        1: 1.2,
        2: 1.0,
        3: 0.9,
        4: 1.1,
        5: 0.5
    }

    weights = [weights_map[tutor_id] for tutor_id in tutor_ids]

    chosen_tutor_id = random.choices(tutor_ids, weights=weights, k=1)[0]

    return chosen_tutor_id
```

```python
assigned_tutor_ids = []

for _, student in df_students.iterrows():
    tutor_id = choose_tutor(student, df_tutors)
    assigned_tutor_ids.append(tutor_id)

df_students["assigned_tutor_id"] = assigned_tutor_ids
df_students.head()
```

## 9. Weekly Slot Assignment

This section assigns weekly lesson slots by matching tutor and student availability.

The aim is to simulate realistic scheduling constraints and identify how well demand can be matched with available teaching capacity.

```python
def assign_weekly_slots(student_row, df_student_available, df_tutor_available):
    student_id = student_row["student_id"]
    tutor_id = student_row["assigned_tutor_id"]
    lessons_per_week = student_row["lessons_per_week"]

    student_slots = set(
        df_student_available[df_student_available["student_id"] == student_id]["slot_key"]
    )

    tutor_slots = set(
        df_tutor_available[df_tutor_available["tutor_id"] == tutor_id]["slot_key"]
    )

    common_slots = list(student_slots.intersection(tutor_slots))

    if len(common_slots) == 0:
        return []

    # sorting
    weekday_order = {
        "Monday": 1,
        "Tuesday": 2,
        "Wednesday": 3,
        "Thursday": 4,
        "Friday": 5,
        "Saturday": 6
    }

    common_slots_sorted = sorted(
        common_slots,
        key=lambda x: (
            weekday_order[x.split("|")[0]],
            x.split("|")[1]
        )
    )

    # random selection – assign as many slots as required
    if len(common_slots_sorted) <= lessons_per_week:
        selected_slots = common_slots_sorted
    else:
        selected_slots = random.sample(common_slots_sorted, lessons_per_week)

    return selected_slots
```

```python
weekly_slots = []

for _, student in df_students.iterrows():
    selected_slots = assign_weekly_slots(student, df_student_available, df_tutor_available)

    for slot in selected_slots:
        weekday, time_block = slot.split("|")

        weekly_slots.append({
            "student_id": student["student_id"],
            "assigned_tutor_id": student["assigned_tutor_id"],
            "weekday": weekday,
            "time_block": time_block,
            "assigned_mode": student["assigned_mode"],
            "typical_duration_minutes": student["typical_duration_minutes"],
            "lessons_per_week": student["lessons_per_week"]
        })

df_weekly_slots = pd.DataFrame(weekly_slots)
df_weekly_slots.head(20)
```

```python
assigned_counts = df_weekly_slots.groupby("student_id").size().reset_index(name="assigned_slots")

df_check = df_students[["student_id", "lessons_per_week"]].merge(
    assigned_counts,
    on="student_id",
    how="left"
)

df_check["assigned_slots"] = df_check["assigned_slots"].fillna(0).astype(int)

df_check.head(20)
```

## 10. Lesson Generation

This section creates the lesson-level dataset.

Each lesson record represents a scheduled teaching event and includes operational, pricing, and delivery-related variables.

This dataset becomes the core fact table for later business analysis.

```python
lesson_rows = []

for _, slot in df_weekly_slots.iterrows():
    student_id = slot["student_id"]
    tutor_id = slot["assigned_tutor_id"]
    weekday = slot["weekday"]
    time_block = slot["time_block"]
    assigned_mode = slot["assigned_mode"]
    duration = slot["typical_duration_minutes"]

    matching_dates = df_calendar_working[df_calendar_working["weekday"] == weekday]

    for _, cal_row in matching_dates.iterrows():
        lesson_rows.append({
            "lesson_date": cal_row["date"],
            "year": cal_row["year"],
            "month": cal_row["month"],
            "month_name": cal_row["month_name"],
            "weekday": weekday,
            "time_block": time_block,
            "student_id": student_id,
            "tutor_id": tutor_id,
            "mode": assigned_mode,
            "duration_minutes": duration
        })

df_lessons = pd.DataFrame(lesson_rows)
df_lessons.head()
```

```python
df_lessons = df_lessons.sort_values(["lesson_date", "time_block", "tutor_id", "student_id"]).reset_index(drop=True)
df_lessons["lesson_id"] = range(1, len(df_lessons) + 1)

df_lessons.head()
```

```python
student_columns = [
    "student_id",
    "student_name",
    "age_group",
    "cefr_level",
    "learning_goal",
    "assigned_tutor_id",
    "assigned_tutor_name",
    "prep_profile",
    "sibling_group_id"
]

df_lessons = df_lessons.merge(
    df_students[student_columns],
    on="student_id",
    how="left"
)

df_lessons.head()
```

```python
tutor_columns = [
    "tutor_id",
    "tutor_name",
    "experience_level",
    "hourly_rate",
    "preferred_mode",
    "home_area"
]

df_lessons = df_lessons.merge(
    df_tutors[tutor_columns],
    on="tutor_id",
    how="left"
)

df_lessons.head()
```

```python
statuses = []

for _, row in df_lessons.iterrows():
    month = row["month"]

    # higher cancellation risk during winter months
    if month in [12, 1, 2]:
        status = random.choices(
            ["completed", "cancelled_by_student", "cancelled_by_tutor", "no_show"],
            weights=[84, 10, 2, 4]
        )[0]
    else:
        status = random.choices(
            ["completed", "cancelled_by_student", "cancelled_by_tutor", "no_show"],
            weights=[88, 7, 2, 3]
        )[0]

    statuses.append(status)

df_lessons["status"] = statuses

df_lessons["status"].value_counts()
```

```python
prep_times = []

for _, row in df_lessons.iterrows():
    prep_profile = row["prep_profile"]

    if prep_profile == "low":
        prep_time = random.choice([5, 10])
    elif prep_profile == "medium":
        prep_time = random.choice([10, 15])
    else:
        prep_time = random.choice([15, 20])

    prep_times.append(prep_time)

df_lessons["prep_time_minutes"] = prep_times

df_lessons["prep_time_minutes"].value_counts().sort_index()
```

```python
travel_times = []

for _, row in df_lessons.iterrows():
    if row["mode"] == "in_person":
        travel_time = random.choice([10, 15, 20, 25, 30, 35, 40])
    else:
        travel_time = 0

    travel_times.append(travel_time)

df_lessons["travel_time_minutes"] = travel_times

df_lessons["travel_time_minutes"].value_counts().sort_index()
```

```python
platforms = []

for _, row in df_lessons.iterrows():
    if row["mode"] == "online":
        platform = random.choices(["Teams", "Skype", "Zoom"], weights=[50, 35, 15])[0]
    else:
        platform = "in_person"

    platforms.append(platform)

df_lessons["platform"] = platforms

df_lessons["platform"].value_counts()
```

```python
df_lessons["base_price"] = (df_lessons["hourly_rate"] * df_lessons["duration_minutes"] / 60).round(2)

df_lessons[["hourly_rate", "duration_minutes", "base_price"]].head()
```

```python
discount_types = []
final_prices = []

for _, row in df_lessons.iterrows():
    base_price = row["base_price"]
    sibling_group_id = row["sibling_group_id"]
    duration = row["duration_minutes"]

    discount_type = "none"
    discount = 0.0

    if pd.notna(sibling_group_id):
        discount_type = "sibling_discount"
        discount = 0.10

    if duration >= 90:
        if discount_type == "none":
            discount_type = "long_lesson_discount"
            discount = 0.08
        else:
            discount_type = "sibling_and_long_lesson"
            discount = 0.15

    final_price = round(base_price * (1 - discount), 2)

    discount_types.append(discount_type)
    final_prices.append(final_price)

df_lessons["discount_type"] = discount_types
df_lessons["final_price"] = final_prices

df_lessons[["base_price", "discount_type", "final_price"]].head(10)
```

## 11. Payments Dataset

This section creates the payments dataset linked to lessons.

It simulates realistic payment outcomes such as:
- paid
- pending
- overdue
- cancelled-related non-payment

This supports analysis of revenue flow and payment behaviour.

```python
payment_rows = []
```

```python
for _, row in df_lessons.iterrows():
    lesson_id = row["lesson_id"]
    lesson_date = row["lesson_date"]
    status = row["status"]
    amount_due = row["final_price"]

    if status == "completed":
        payment_status = random.choices(
            ["paid", "late", "missing"],
            weights=[88, 9, 3]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "late":
            amount_paid = amount_due
        else:
            amount_paid = 0

    elif status == "cancelled_by_student":
        payment_status = random.choices(
            ["paid", "partial", "missing"],
            weights=[20, 35, 45]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "partial":
            amount_paid = round(amount_due * 0.5, 2)
        else:
            amount_paid = 0

    elif status == "cancelled_by_tutor":
        payment_status = "waived"
        amount_paid = 0

    else:  # no_show
        payment_status = random.choices(
            ["paid", "partial", "missing"],
            weights=[45, 25, 30]
        )[0]

        if payment_status == "paid":
            amount_paid = amount_due
        elif payment_status == "partial":
            amount_paid = round(amount_due * 0.5, 2)
        else:
            amount_paid = 0

    payment_rows.append({
        "lesson_id": lesson_id,
        "lesson_date": lesson_date,
        "amount_due": amount_due,
        "amount_paid": amount_paid,
        "payment_status": payment_status
    })

df_payments = pd.DataFrame(payment_rows)
df_payments.head()
```

```python
df_payments["payment_id"] = range(1, len(df_payments) + 1)
df_payments.head()
```

```python
payment_dates = []

for _, row in df_payments.iterrows():
    lesson_date = pd.to_datetime(row["lesson_date"])
    payment_status = row["payment_status"]

    if payment_status == "paid":
        offset_days = random.choice([0, 1, 2, 3])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    elif payment_status == "late":
        offset_days = random.choice([5, 7, 10, 14])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    elif payment_status == "partial":
        offset_days = random.choice([0, 2, 5, 7])
        payment_date = lesson_date + pd.Timedelta(days=offset_days)

    else:
        payment_date = pd.NaT

    payment_dates.append(payment_date)

df_payments["payment_date"] = payment_dates
df_payments.head()
```

```python
df_payments = df_payments[
    [
        "payment_id",
        "lesson_id",
        "lesson_date",
        "amount_due",
        "amount_paid",
        "payment_status",
        "payment_date"
    ]
]

df_payments.head()
```

```python
expected_revenue = df_lessons["final_price"].sum()
actual_revenue = df_payments["amount_paid"].sum()
revenue_leakage = expected_revenue - actual_revenue

print("Expected revenue:", round(expected_revenue, 2))
print("Actual revenue:", round(actual_revenue, 2))
print("Revenue leakage:", round(revenue_leakage, 2))
```

```python
print("Total lessons:", len(df_lessons))
print()
print("Lesson status breakdown:")
print(df_lessons["status"].value_counts())
print()
print("Mode breakdown:")
print(df_lessons["mode"].value_counts())
```

```python
cancelled_lessons = df_lessons["status"].isin(["cancelled_by_student", "cancelled_by_tutor", "no_show"]).sum()
total_lessons = len(df_lessons)
cancellation_rate = cancelled_lessons / total_lessons

print("Cancelled / not completed lessons:", cancelled_lessons)
print("Cancellation / disruption rate:", round(cancellation_rate * 100, 2), "%")
```

```python
tutor_workload = df_lessons.groupby(["tutor_id", "tutor_name"]).agg(
    total_lessons=("lesson_id", "count"),
    total_teaching_minutes=("duration_minutes", "sum"),
    avg_price=("final_price", "mean")
).reset_index()

tutor_workload["total_teaching_hours"] = (tutor_workload["total_teaching_minutes"] / 60).round(2)

tutor_workload.sort_values("total_lessons", ascending=False)
```

```python
tutor_revenue = df_lessons.groupby(["tutor_id", "tutor_name"]).agg(
    expected_revenue=("final_price", "sum")
).reset_index()

actual_by_tutor = df_lessons.merge(
    df_payments[["lesson_id", "amount_paid"]],
    on="lesson_id",
    how="left"
).groupby(["tutor_id", "tutor_name"]).agg(
    actual_revenue=("amount_paid", "sum")
).reset_index()

tutor_revenue_summary = tutor_revenue.merge(actual_by_tutor, on=["tutor_id", "tutor_name"])
tutor_revenue_summary["revenue_gap"] = (
    tutor_revenue_summary["expected_revenue"] - tutor_revenue_summary["actual_revenue"]
).round(2)

tutor_revenue_summary.sort_values("expected_revenue", ascending=False)
```

```python
age_group_summary = df_lessons.groupby("age_group").agg(
    total_lessons=("lesson_id", "count"),
    avg_price=("final_price", "mean"),
    total_revenue=("final_price", "sum")
).reset_index()

age_group_summary.sort_values("total_revenue", ascending=False)
```

```python
mode_summary = df_lessons.groupby("mode").agg(
    total_lessons=("lesson_id", "count"),
    avg_price=("final_price", "mean"),
    total_revenue=("final_price", "sum"),
    avg_travel_time=("travel_time_minutes", "mean")
).reset_index()

mode_summary
```

```python
payment_summary = df_payments.groupby("payment_status").agg(
    count=("payment_id", "count"),
    total_due=("amount_due", "sum"),
    total_paid=("amount_paid", "sum")
).reset_index()

payment_summary
```
